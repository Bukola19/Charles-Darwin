```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
fig1b = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vTWWyFhXK03nG8pUbG8dcsXH5SiPXikXd-9bN5NcYcyYbPkPkmBNIGypZppIgViVysHBDRsW7JXlt7B/pub?gid=1700050588&single=true&output=csv')
fig1c = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vTWWyFhXK03nG8pUbG8dcsXH5SiPXikXd-9bN5NcYcyYbPkPkmBNIGypZppIgViVysHBDRsW7JXlt7B/pub?gid=1700799716&single=true&output=csv')
```

```python
average_df = fig1b.groupby(['grna', 'Day'])['Rate'].mean().reset_index()

# Normalize the rates
max_scale = lambda x: x / x.max()
average_df['Normalized_Rate'] = average_df.groupby('grna')['Rate'].transform(max_scale)

# Define a color palette
beige_palette = palette=['#305695', '#4474c5', '#8facdd']


# Plot the data
sns.barplot(data=average_df, x='grna', y='Normalized_Rate', hue='Day', palette=beige_palette, order=['CCR5', 'BRCA1 #1', 'BRCA1 #2'])




plt.title('')
plt.xlabel(None)
plt.yticks([0, 0.5, 1], [0, 0.5, 1])
plt.ylabel('Relative Indel Frequency')
plt.legend(title='Day')
plt.show()
```
